```python
#1st program

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


data={
    'total_bill':[10,20,30,40],
    "tip":[1,2,3,4],
    "smoker":["yes","no","yes","no"],
    "time":["lunch","dinner","lunch","dinner"]
    }


ds=pd.DataFrame(data)

#scatterplot
sns.set(style="darkgrid")
sns.relplot(x="total_bill",y="tip",data=ds)
sns.relplot(x="total_bill",y="tip",hue="smoker",data=ds)
sns.relplot(x="total_bill",y="tip",hue="smoker",col="time",data=ds)

#correlation heatmap

ds_numeric = ds.copy()
ds_numeric['smoker'] = ds_numeric['smoker'].map({'yes': 1, 'no': 0})
ds_numeric['time'] = ds_numeric['time'].map({'lunch': 1, 'dinner': 0})

plt.matshow(ds_numeric.corr())
plt.title("correlation heatmap")
plt.colorbar()
plt.show()
```

```python
#3rd program-matrices and tables

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

ds=pd.DataFrame({"category":['A','B','C','D','E','F'],
                 "Subcategory":["x","y","x","y","x","y"],
                 "sales":[100,150,200,250,300,350],
                 "Profit":[20,30,50,60,80,90]})
print(ds)

matrix=ds.pivot_table(values='sales',index="category",columns='Subcategory',aggfunc="sum",fill_value=0)

print(matrix)

plt.figure(figsize=(6,4))
sns.heatmap(matrix, annot=True, fmt=".0f", cmap="coolwarm",linewidths=0.5)
plt.title("sales matrix(category vs Subcategory)")
plt.show()
```

```python
#Ex-4 slicers and map visualization

import pandas as pd
import folium
from IPython.display import display
# Sample dataset (Power BI will replace this with actual data)
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami'],
    'Region': ['East', 'West', 'Midwest', 'South', 'East'],
    'Latitude': [40.7128, 34.0522, 41.8781, 29.7604, 25.7617],
    'Longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -80.1918],
    'Sales': [500, 700, 400, 600, 550]   }
df = pd.DataFrame(data)
#  Slicer Feature: Filter Data (Example: Show only 'East' Region)
selected_region = 'East'  # You can change this dynamically using a slicer in Power BI
filtered_df = df[df['Region'] == selected_region]
#  Map Visualization using Folium
map = folium.Map(location=[39.8283, -98.5795], zoom_start=4)  # USA Centered
# Add points to the map
for index, row in filtered_df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['City']} - Sales: ${row['Sales']}",
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(map)
display(map)
```

```python
#Ex-5-Gauges and Single Number Cards
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Sample dataset (Replace this with Power BI data)
data = {
    'Metric': ['Total Sales'],
    'Value': [750],
    'Target': [1000]
}
df = pd.DataFrame(data)
# Extract values for visualization
actual_value = df['Value'][0]
target_value = df['Target'][0]
#  **Single Number Card**: Display Key Metric
print(f"Total Sales: ${actual_value}")
#  **Gauge Chart Visualization**
fig, ax = plt.subplots(figsize=(6, 3))
# Create Gauge Chart
gauge_colors = ['red' if actual_value < target_value * 0.5 else
                'orange' if actual_value < target_value * 0.8 else
                'green']
ax.barh(0, actual_value, color=gauge_colors, left=0, height=0.4)
ax.barh(0, target_value, color='gray', left=actual_value, height=0.4)
ax.set_xlim(0, target_value * 1.2)
ax.set_yticks([])
ax.set_xticks([0, target_value * 0.5, target_value * 0.8, target_value])
ax.set_xticklabels(['0%', '50%', '80%', '100%'])
ax.set_title("Sales Performance Gauge")
plt.show()
```

```python
#Ex-6-Modifying Colors in Charts and Visuals
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample dataset (Power BI will replace this with actual data)
data = {
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Toys', 'Books'],
    'Sales': [800, 300, 500, 200, 1000]    }
df = pd.DataFrame(data)
#  **Custom Color Logic** for Bar Chart
def get_color(sales):
    if sales < 300:
        return 'red'  #  Low Sales
    elif sales < 700:
        return 'orange'  #  Medium Sales
    else:
        return 'green'  #  High Sales
df['Color'] = df['Sales'].apply(get_color)
#  **Bar Chart with Dynamic Colors**
plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Sales'], color=df['Color'])
plt.xlabel("Product Category")
plt.ylabel("Sales Amount")
plt.title("Sales by Category (with Dynamic Colors)")
plt.show()
#  **Heatmap with Custom Color Palette**
plt.figure(figsize=(6, 4))
heatmap_data = pd.DataFrame({'Sales': df['Sales']}, index=df['Category'])
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="coolwarm", linewidths=0.5)
plt.title("Sales Heatmap")
plt.show()


```

```python
#ex=9==Page Layout and Formatting

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample dataset (Power BI will replace this with actual data)
data = {
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Toys', 'Books'],
    'Sales': [800, 300, 500, 200, 1000],
    'Profit': [120, 50, 80, 30, 200]
}
df = pd.DataFrame(data)
#  **Create a Figure with Multiple Charts (Grid Layout)**
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns
#  **First Chart: Sales Bar Chart**
sns.barplot(x='Category', y='Sales', data=df, palette="Blues", ax=axes[0])
axes[0].set_title("Sales by Category")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Sales Amount")
#  **Second Chart: Profit Line Chart**
sns.lineplot(x='Category', y='Profit', data=df, marker='o', color='green', ax=axes[1])
axes[1].set_title("Profit by Category")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Profit Amount")
#  **Formatting & Layout Adjustments**
plt.tight_layout()  # Optimizes spacing
plt.show()

```
